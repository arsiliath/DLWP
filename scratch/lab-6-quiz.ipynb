```python
# so the function is

def compute_output_shape(B, K, W, S, padding, filters):
    if padding == 'valid':
        P = 0
        # this is more complex is S is higher?
    else: # same
        P = (K - 1) // 2
        # this is more complex is S is higher?

    out_size = (W - K + 2 * P) // S + 1 # floor div

    return (B, out_size, out_size, filters)
    
```
